## Python Basics

### 1. Basic Data Types
Python supports the following basic data types:
- int (Integer numbers)
- float (Decimal numbers)
- complex (Complex numbers)
- str (String)
- bool (Boolean)

````
a = 10
b = 3.14
c = 2+3j
d = "Python"
e = True

type(a), type(b), type(c), type(d), type(e)
````

### 2. type() Function (Implementation)
The `type()` function is used to find the data type of a variable.

````
x = 25
print(type(x))

y = 4.5
print(type(y))
````

### 3. Input Function
The `input()` function is used to take input from the user. It always returns a string.

````
name = input("Enter your name: ")
print("Hello", name)
print(type(name))
````

### 4. Working of type() Function
The `type()` function checks the class to which a variable belongs and returns its data type.

````
value = 100
print("Value:", value)
print("Data type:", type(value))
````

### 5. Need of Typecasting with Program
Typecasting is required to convert one data type into another.

````
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")

sum = int(num1) + int(num2)
print("Sum:", sum)
````

### 6. Function Syntax
Functions are used to divide a program into reusable blocks.

**Syntax:**
```
def function_name(parameters):
    statements
    return value
```

### 7. Rectangular Area Program

````
def rectangle_area(length, breadth):
    return length * breadth

l = int(input("Enter length: "))
b = int(input("Enter breadth: "))
print("Area of Rectangle:", rectangle_area(l, b))
````
