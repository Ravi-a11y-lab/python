41. Write Pandas code to merge two datasets using inner and outer joins.

```python
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'id': [2, 3, 4], 'score': [85, 90, 88]})

# Inner Join (keeps only matching IDs)
inner_merged = pd.merge(df1, df2, on='id', how='inner')

# Outer Join (keeps all IDs from both, filling missing matches with NaN)
outer_merged = pd.merge(df1, df2, on='id', how='outer')
```

49. Write code to reduce memory usage of a DataFrame.

```python
def optimize_memory(df):
    for col in df.columns:
        if df[col].dtype == 'float64':
            df[col] = pd.to_numeric(df[col], downcast='float')
        elif df[col].dtype == 'int64':
            df[col] = pd.to_numeric(df[col], downcast='integer')
        elif df[col].dtype == 'object' and df[col].nunique() / len(df) < 0.5:
            df[col] = df[col].astype('category')
    return df
```

55. Implement RFM (Recency, Frequency, Monetary) analysis using Pandas.

```python
import datetime as dt

# Assume df has columns: 'customer_id', 'order_date', 'total_price'
NOW = dt.datetime(2023, 12, 31)
df['order_date'] = pd.to_datetime(df['order_date'])

rfm = df.groupby('customer_id').agg({
    'order_date': lambda x: (NOW - x.max()).days, # Recency
    'customer_id': 'count',                       # Frequency
    'total_price': 'sum'                          # Monetary
}).rename(columns={'order_date': 'Recency', 'customer_id': 'Frequency', 'total_price': 'Monetary'})
```

56. Compute Customer Lifetime Value (CLV).

```python
# Average Order Value (AOV)
aov = rfm['Monetary'].sum() / rfm['Frequency'].sum()

# Purchase Frequency (PF)
purchase_freq = rfm['Frequency'].sum() / len(rfm)

# Average Customer Lifespan (Assumed 3 years for this example)
lifespan = 3

# CLV Calculation
clv = aov * purchase_freq * lifespan
print(f"Estimated CLV: ${clv:.2f}")
```

58. Design a simple end-to-end data pipeline using Python

```python
def extract(file_path):
    return pd.read_csv(file_path)

def transform(df):
    df = df.dropna()
    df['date'] = pd.to_datetime(df['date'])
    df['revenue'] = df['qty'] * df['price']
    return df

def load(df, db_name, table_name):
    conn = sqlite3.connect(db_name)
    df.to_sql(table_name, conn, if_exists='append', index=False)

# Execution
def run_pipeline():
    raw_data = extract('daily_sales.csv')
    clean_data = transform(raw_data)
    load(clean_data, 'company_db.sqlite', 'sales_fact')

run_pipeline()
```
